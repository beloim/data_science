```python
#Convertendo imagem para escala de cinza
from skimage.io import imread
from skimage.transform import resize

import matplotlib.pyplot as plt
import matplotlib.cm as cm

colorida=imread('https://softgraf.com/img/img2.jpg')
print(colorida.ndim,colorida.shape)

cinza=imread('https://softgraf.com/img/img2.jpg', as_gray=True)
print(cinza.ndim,cinza.shape)

plt.imshow(colorida)
plt.show()

plt.imshow(cinza)
plt.show()

plt.imshow(cinza,cmap=cm.gray)
plt.show()

plt.imshow(cinza,cmap=cm.plasma)
plt.show()

plt.imshow(cinza,cmap=cm.magma)
plt.show()

plt.imshow(cinza,cmap=cm.inferno)
plt.show()

#mapas de cores
#https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
#cortando uma imagem
recortada = colorida[0:200,0:500] #linha 0 a 199 ,coluna 0 a 500
plt.imshow(recortada)
plt.show()
```

```python
#redimensionando a imagem(abaixa a resoluçao,pois nao muda o tamanho)
print(colorida.shape) #(390,640)

redim=resize(colorida,(39,64))
print(redim.shape)
plt.imshow(redim)
plt.show()
```

```python
#Convertendo matriz da imagem para uma linha unica=3D para 1D
img_row= colorida.flatten()
print(img_row.ndim,img_row.shape)
```

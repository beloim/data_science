#### imporrtaçao/exportaçao para banco de dados SQLite

```python
import sqlite3
import pandas as pd
```

```python
#importando CSV
df=pd.read_csv('https://softgraf.com/cursodatascience/dados-ibge-2010.csv')
df
```

```python
#cria um banco de dados 'teste_banco.slite' na pasta atual dp notebook
#e abre a conexao(comunicaçao) com o banco
conexao=sqlite3.connect('teste_banco.sqlite')
```

```python
#grava todos os dados no dataframe na 'tabela_ibge'
df.to_sql(con=conexao,name='tabela_ibge',if_exists='replace',index=False)
#fecha conexao
conexao.close()
```

```python
#AGORA VOU IMPORTAR OS DADOS DO BANCO
#abra a conexao com o banco
conexao=sqlite3.connect('teste_banco.sqlite')

#traz todos os dados da tabela e converte em um dataframe
df=pd.read_sql_query('SELECT * FROM tabela_ibge',conexao)
display(df)

#fecha conexao
conexao.close()
```

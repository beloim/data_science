#### Criando um data frame a partir de um array numpy

```python
import numpy as np

#arange = cria um array entre 1 e 25
#reshape = converte em um array de 5 linhas x 5 colunas
a = np.arange(1,26).reshape((5,5))
display(a)

#cria um data frame com indice e rotulos padroes
df = pd.DataFrame(a)
display(df)
```

```python
#criando o mesmo data frame, porem agora com indices e rotulos personalizados
df2 = pd.DataFrame(a,
                  columns=['a','b','c','d','e'],
                  index=['x','y','z','w','v']
                  )
#df2.columns=['c1','c2','c3','c4','c5']
display(df2)
print(df2.columns)
print(df2.index)
#imprime os dados da 'linha' (indice) = x
print(df2.loc['x'])
print(type(df2.loc['x']))
```

```python
### criando uma serie a partir de uma lista
dias=['domingo','segunda','terca','quarta','quinta','sexta','sabado']
s=pd.Series(dias)
#personaliza o indixe da serie
s.index=['dia1','dia2','dia3','dia4','dia5','dia6','dia7']

display(s)
print(s['dia5'])
```

### Trocando o indice numerico padrao por uma coluna em uma data frame

```python
import numpy as np

#arange = cria um array entre 1 e 25
#reshape = converte em um array de 5 linhas x 5 colunas
a = np.arange(1,26).reshape((5,5))


#cria um data frame com 5 linhas x 5 colunas e personaliza as colunas
#personaliza as colunas
df = pd.DataFrame(a,columns=['a','b','c','d','e'])
display(df)
#troca o indice atual (0,1,2,3,4) pela coluna a (1,6,11,16,21)
df=df.set_index('a')
display(df)
#imprime a mesma linha do data frame
print(df.loc[21]) #usa o indice
print(df.iloc[4]) #usa a posiçao
type(df.iloc[4])
```

### Obtendo series com os operadores loc e iloc

```python
a= np.arange(1,10).reshape((3,3))
df=pd.DataFrame(a,index=['x','y','z'],columns=['A','B','C'])
display(df)

#acessando as linhas com operador iloc[]
#retorna uma serie para cada lina
print(df.iloc[0])
print(df.iloc[1])
print(df.iloc[2])

#acessando as linhas com operador loc[]
print(df.loc['x'])
print(df.loc['y'])
print(df.loc['z'])

#converte uma serie en lista
print(list(df.loc['x']))

#converte uma serie em tupla
print(tuple(df.loc['x']))

#converte uma serie em conjunto
print(set(df.loc['x']))

#converte uma serie em dicionario
print(dict(df.loc['x']))
```

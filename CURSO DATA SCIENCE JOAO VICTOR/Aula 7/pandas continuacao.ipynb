```python
import pandas as pd

#variavel categorica qualitativa nominal
cores= ['azul','vermelho','verde']

#cria uma serie do pandas com o tipo 'category'
serie_cor =pd.Series(['azul','vermelho','verde'],dtype='category')
print(serie_cor)
#cria uma variavel categorica a partir de grupo de cores
categorica=pd.Categorical(['azul','vermelho','verde','branco','amarelo'],categories=cores)
print('\ncategorica', categorica)

print('\nMostra valores nulos:',categorica.isnull())

print('\nElimina is nulos:',categorica.dropna())
```

```python
#cria uma serie a partir da variavel categorica
dados = pd.Series(categorica)
print(type(dados))
display(dados)

#cria um dataframe a partir da serie
#os valores que nao fazem parte da categoria aparecem como NaN
df=pd.DataFrame(dados)
display(df)

#mostra valores NaN = True
display(pd.isnull(df))

#filtra somente os valores NaN
pesquisa=pd.isnull(dados)
display(pesquisa[pesquisa==True])
```

#### validando os dados no dataframe - dropna(),duplicated(),drop_duplicates()

```python
import pandas as pd

df=pd.DataFrame(columns=['bebida','tipo','calorias','açucar(g)','cafeina(mg)'])
df['bebida']=['cafe','cafe','cafe com leite','cafe mocha','cappucino','cafe','cafe gelado']
#cria uma serie categorica
tipo=pd.Series(['quente','gelado'],dtype='category')
#adiciona a coluna tipo,porem com uma variavel categorica
df['tipo']=pd.Categorical(['quente','quente','quente','quente','quente','morno','gelado'],categories=tipo)

df['calorias']=[4,4,100,170,60,30,60]
df['açucar(g)']=[0,0,14,27,8,0,15]
df['cafeina(mg)']=[260,260,75,95,75,50,120]
display(df)
#no indice 5 do dataframe tempos um valor categorico invalido,entao vamos eliminar
df.dropna(inplace=True) #inplace permite salvar na propria variavel os dados diferentes
display(df)

#no indice 1, tempos valores duplicados,entao vamos eliminar
#encontra as linhas duplicadas,craindo uma serie com valores True e False
pesquisa=pd.DataFrame.duplicated(df)
display(pesquisa)
#mostra somente True
print(pesquisa[pesquisa==True])
#agora elimina
df.drop_duplicates(inplace=True)
#dataframe sem linhas duplicadas
display(df)

#apos eliminar a linha 1,precisamos restar
#drop=True -> elimina o indice anterior e cria um novo
df.reset_index(inplace=True,drop=True)
display(df)
```

#### tratando dados ausentes (NaN) com filina() e dropna()

#### Filtrando dados com pandas.DataFrame.where()

```python
import pandas as pd
df=pd.read_excel('https://softgraf.com/cursodatascience/bebida.xlsx')
display(df)

#se as calorisas forem iguais ou superiores a 100
df2=df.where(df.Calorias >= 100)
display(df2)


#elimina os nulos
display(df2.dropna().reset_index(drop=True))

#substituindo os dados nulos por outro valor
df2=df.where(df.Calorias >= 100, other=' ')
display(df2)

#se a bebida nao tiver açucar 
df2=df.where(df['Açúcar (g)'] == 0, other=' ')
display(df2)

#se a bebida iniciar por Café & Calorías entre 60 e 100 & Cafeína menor que 100
cond1=df.Bebida=='Café'
cond2=df.Bebida=='Café com leite'
cond3=df.Bebida=='Café Mocha'
cond4=df.Bebida=='Café Gelado'

#resume as condiçoes acima em uma unica
#OU == |
#E  == &
cond5=cond1|cond2|cond3|cond4
cond6=(df.Calorias >= 60) & (df.Calorias <= 100)
cond7=df['Cafeína (mg)'] < 100

#agora faz o filtro com where
df2=df.where(cond5 & cond6 & cond7)
display(df2.dropna().reset_index(drop=True))
```

#### filtrando dados com pandas.DataFrame.filter()

```python
df=pd.read_excel('https://softgraf.com/cursodatascience/bebida.xlsx')
#filtra colunas pelo label (nome da coluna)
df2=df.filter(items=['Tipo','Calorias'])
display(df2)

#porem posso fazer sem o filtro
df=df[['Tipo','Calorias']]
display(df2) #mesmo resultado

#filtra colunas contedo 'Café' (pega a coluna Cafeína)
df2=df.filter(like='Caf',axis=1) #1=coluna
display(df2)

#usando expressao regular
#todos contendo a sequencia "Ca"
df2=df.filter(regex='Ca',axis=1)
display(df2)
```

#### filtrando dados com operadores logicos

```python
#filtra pelo valor das Calorias
df=pd.read_excel('https://softgraf.com/cursodatascience/bebida.xlsx')
df2=df[df.Calorias >60]
display(df2)

#Filtra pelo valor da Cafeína
df2=df[df['Cafeína (mg)']>100]
display(df2)

#filtra pela cafeína ou calorias
df2=df[(df['Cafeína (mg)']>100)|(df.Calorias>60)]
display(df2)
```

#### Filtrando dados com pandas.DataFrame.isin()

```python
#filtra pelo nome da Bebufa usando uma lista
bebidas=['Café Mocha','Café Gelado']
df2=df[df.Bebida.isin(bebidas)]
display(df2)

#filtra todos os cafes gelados
tipos=['Gelado']
df2=df[(df.Bebida.isin(bebidas))&(df.Tipo.isin(tipos))]
display(df2)
```

#### Filtrando dados com base em strings

```python
#filtra Bebida iniciando com 'Cap'
df2=df[df.Bebida.str.startswith('Cap')]
display(df2)

#filtra Tipo terminando em 'do'
df2=df[df.Tipo.str.endswith('do')]
display(df2)

#filtra Bebida contendo LEITE
df2=df[df.Bebida.str.contains('LEITE',case=False)]
display(df2)

#filtra Bebida contendo 'Leite quente'
df2=df[(df.Bebida.str.contains('leite')) & (df.Tipo.str.contains('quente',case=False))]
display(df2)

#filtra bebidas que nao contem leite
df2=df[~df.Bebida.str.contains('leite')] #colocar um tio '~' pede o contrario do enunciado
display(df2)
```
